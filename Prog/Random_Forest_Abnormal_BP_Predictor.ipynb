```python
#In this code, Random Forest is implemented on Blood pressure abnomality data of patients.

import pandas as pd
```

```python
#checking the null values in our data set
bp_data.isnull().sum()
```

```python
#data cleaning
bp_data.Genetic_Pedigree_Coefficient=bp_data.Genetic_Pedigree_Coefficient.fillna(0)
bp_data.Pregnancy = bp_data.Pregnancy.fillna(0)
bp_data.alcohol_consumption_per_day = bp_data.alcohol_consumption_per_day.fillna(0)
```

```python
#Describing data

#bp_data = bp_data.iloc[:,1:15]

bp_data.describe(include='all')
```

```python
#data pre processing. Converting features into categorical features

bp_data.Blood_Pressure_Abnormality= bp_data.Blood_Pressure_Abnormality.astype('category')
bp_data.Sex = bp_data.Sex.astype('category')
bp_data.Smoking = bp_data.Smoking.astype('category')
bp_data.Pregnancy = bp_data.Pregnancy.astype('category')
bp_data.Level_of_Stress = bp_data.Level_of_Stress.astype('category')
bp_data.Chronic_kidney_disease = bp_data.Chronic_kidney_disease.astype('category')
bp_data.Adrenal_and_thyroid_disorders = bp_data.Adrenal_and_thyroid_disorders.astype('category')
```

```python
#importing random forest classifier from sklearn library

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
#splitting data into target variable and independent feature

bp_data_X = bp_data.iloc[:,2:15]
bp_data_Y = bp_data.iloc[:,1]

bp_data_X.head()
```

```python
#splitting data into training and testing

bp_data_X_train,bp_data_X_test, bp_data_Y_train, bp_data_Y_test = train_test_split(bp_data_X,bp_data_Y,test_size=0.2,random_state=80)
```

```python
# creating instance of random forest classifier. Taking n_estimators (number of trees to be their in our forest) = 100
rndm_fr = RandomForestClassifier(n_estimators=100)
```

```python
#using fit method, training our model on train data.
rndm_fr.fit(bp_data_X_train,bp_data_Y_train)
```

```python
#predicting values for the test data x on our model.
predic_value= rndm_fr.predict(bp_data_X_test)
```

```python
predic_value
```

```python
#plotting the confusion matrix for the predicted values by our model.

conf_mat = confusion_matrix(predic_value, bp_data_Y_test)
conf_mat
```

```python
#calculating the accuracy through confusion matrix.

bp_acc=conf_mat.diagonal().sum()/conf_mat.sum()*100
bp_acc
```
